# Exploratory Analysis\n
Use this notebook to inspect cleaned and integrated allied health workforce datasets.

```python
import pandas as pd\n
import matplotlib.pyplot as plt\n
\n
workforce = pd.read_csv('../data/processed/workforce_clean.csv')\n
population = pd.read_csv('../data/processed/population_clean.csv')\n
workforce.head()
```

```python
summary = workforce.groupby(['year', 'region'], as_index=False)['workforce_supply'].sum()\n
pivot = summary.pivot(index='year', columns='region', values='workforce_supply')\n
pivot.plot(figsize=(10, 6), marker='o')\n
plt.title('Workforce Supply by Region')\n
plt.xlabel('Year')\n
plt.ylabel('Supply')\n
plt.grid(True, alpha=0.3)\n
plt.show()
```
